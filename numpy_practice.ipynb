```python
# Create a 1D array of daily sales figures (in thousands)
daily_sales = np.array([15.2, 18.5, 14.0, 20.1, 19.8]) 
print(f"Original Sales: {daily_sales}")

# 1. Add a constant (100) to every element
bonus_added = daily_sales + 100
print(f"Sales + 100 Bonus: {bonus_added}")

# 2. Multiply every element by a constant factor (e.g., a 10% commission rate)
commission = daily_sales * 0.10
print(f"10% Commission: {commission}")

# 3. Square every element (using the ** operator)
squared_sales = daily_sales ** 2
print(f"Sales Squared: {squared_sales}")
```

```python
# A 3x2 Matrix (3 products, 2 regions)
sales_matrix = np.array([
    [100, 150],
    [200, 250],
    [50, 75]
])
# Shape: (3, 2)

# A 1D array representing a discount rate for each region
# Region 1 gets 10% off (0.9 multiplier), Region 2 gets 5% off (0.95 multiplier)
discount_rate = np.array([0.9, 0.95]) 
# Shape: (2,)

print("Sales Matrix:")
print(sales_matrix)
print(f"\nDiscount Rates (Shape: {discount_rate.shape}): {discount_rate}")

# When we multiply, the 1D array (shape 2) is broadcast across the rows of the 3x2 matrix.
# [0.9, 0.95] is effectively copied three times:
# [[0.9, 0.95],
#  [0.9, 0.95],
#  [0.9, 0.95]]
final_sales = sales_matrix * discount_rate

print("\nFinal Sales (Sales * Discount Rate):")
print(final_sales)
```

```python
# 1D array of 12 monthly sales figures
monthly_sales = np.arange(101, 113) # Creates [101, 102, ..., 112]
print(f"Original 1D array (Size {monthly_sales.size}): {monthly_sales}")

# Reshape the 1D array (size 12) into a 2D array of 4 rows and 3 columns
quarterly_sales = monthly_sales.reshape(4,3) 

print("\nReshaped 4x3 Quarterly Sales Matrix:")
print(quarterly_sales)
print(f"New Shape: {quarterly_sales.shape}")
```

```python
# Original array: monthly_sales (size 12)

# Reshape into a matrix with 3 columns, let NumPy calculate the rows (-1)
three_column_matrix = monthly_sales.reshape(-1, 3) 
print("\nReshaped with -1 (3 columns, 4 rows calculated):")
print(three_column_matrix)
print(f"New Shape: {three_column_matrix.shape}")

# Reshape into a matrix with 6 rows, let NumPy calculate the columns (-1)
six_row_matrix = monthly_sales.reshape(6, -1)
print("\nReshaped with -1 (6 rows, 2 columns calculated):")
print(six_row_matrix)
print(f"New Shape: {six_row_matrix.shape}")
```

# NumPy Basics - Introduction to Arrays

## What is NumPy?
NumPy (Numerical Python) is a fundamental library for scientific computing in Python. It provides:
- Powerful N-dimensional array objects (ndarray)
- Fast operations on arrays
- Tools for working with numerical data

## Why Use NumPy Arrays?
- **Performance**: NumPy arrays are much faster than Python lists for numerical operations
- **Memory Efficiency**: Uses less memory than Python lists
- **Convenience**: Provides many mathematical functions and operations

## Key Concept: ndarray
The core of NumPy is the ndarray (n-dimensional array) - a homogeneous collection of elements of the same data type.

## Importing NumPy Library
Before we can use NumPy, we need to import it into our Python program. By convention, we import NumPy with the alias 'np' for convenience.

```python
# Import the NumPy library and give it the alias 'np'
# This makes it easier to reference NumPy functions throughout our code
import numpy as np
```

## Creating a NumPy Array from a List
Now we'll create our first NumPy array using the `np.array()` function. This function converts a Python list or tuple into a NumPy array (ndarray).

```python
# Create a NumPy array from a Python list containing integers
a1 = np.array([6, 2, 8, 11, 9])  # np.array() converts the list to a NumPy array
a1  # Display the created array
```

## Checking the Data Type of a NumPy Array
We can use Python's `type()` function to verify that our variable is indeed a NumPy array. This will return `numpy.ndarray`.

```python
# Check the data type of the variable a1
# The type() function returns the class type of the object
type(a1)  # This will output: numpy.ndarray
```

## Creating a Python List and Converting to Array
We can also store our data in a Python list variable first, and then convert that list to a NumPy array. This two-step approach is useful when working with data from other sources.

```python
# Create a Python list with integer values
list1 = ([11, 22, 33, 44])  # Store the list in a variable called list1
list1  # Display the list
```

## Converting an Existing List Variable to NumPy Array
Now we take the list variable we created (`list1`) and convert it to a NumPy array using `np.array()`.

```python
# Pass the list variable (list1) to np.array() function
# This converts the Python list to a NumPy array
np.array(list1)  # Returns a NumPy array with values [11, 22, 33, 44]
```

## Creating an Array with Float (Decimal) Values
NumPy arrays can hold different data types. Here we create an array with floating-point (decimal) numbers.

```python
# Create a NumPy array with floating-point (decimal) values
# np.array() automatically detects the data type as float
np.array([1.1, 1.2, 1.3])  # Returns an array of type float64
```

## Automatic Type Conversion (Upcasting) - Mixed Integer and Float
When you create an array with mixed data types (integers and floats), NumPy automatically converts all values to the "higher" data type. In this case, integers are converted to floats to maintain consistency.

```python
# Create an array with mixed data types: floats (1.1, 1.3) and integers (2, 4)
# NumPy automatically converts integers to floats (upcasting)
np.array([1.1, 2, 1.3, 4])  # Result: [1.1, 2.0, 1.3, 4.0] - all floats
```

```python
# Create an array with mixed data types: float, int, and string
# NumPy converts ALL elements to the same type (strings in this case)
# String is the 'highest order' type that can represent all values
np.array([1.1, 2, 'hi'])  # All elements become strings: ['1.1', '2', 'hi']
```

## Creating Arrays with np.arange()

The `np.arange()` function creates an array with a sequence of numbers. It takes three parameters:
- **start**: The starting value (inclusive)
- **stop**: The ending value (exclusive - not included in the array)
- **step**: The difference between each consecutive number

This is similar to Python's built-in `range()` function, but creates a NumPy array instead of a list.

```python
# Create an array starting from 20, up to (but not including) 50, with a step of 2
# This will generate: [20, 22, 24, 26, 28, 30, 32, 34, 36, 38, 40, 42, 44, 46, 48]
arr1 = np.arange(20, 50, 2)
```
