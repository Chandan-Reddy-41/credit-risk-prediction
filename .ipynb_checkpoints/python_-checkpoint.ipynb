```python
# 1. Define the calculator function (no parameters needed)
def calculator():
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    result = num1 + num2
    print(f"The result is: {result}")
    return result

# 2. Use a while loop to control the program flow
continue_calculating = True

while continue_calculating:
    # Call the calculator function to perform the task
    calculator()

    # Ask the user if they want to continue
    response = input("Do you want to calculate again? (yes/no): ").lower()

    # Check the response to decide whether to continue the loop
    if response == 'no' or response == 'n':
        continue_calculating = False

print("Calculator finished.")
```
