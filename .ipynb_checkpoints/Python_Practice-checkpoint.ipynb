##### **heading tags**

A single hash symbol (`#`) at the beginning of a line creates a heading in Markdown. The number of hashes determines the size of the heading.

# Heading 1 - The largest and most prominent heading.
## Heading 2 - A smaller subheading.
### Heading 3 - Even smaller.
#### Heading 4 - A smaller heading.
##### Heading 5 - A smaller heading.
###### Heading 6 - The smallest heading.
This system lets you create a logical hierarchy in your documents, much like a book has chapters, sections, and subsections.

##### **Identifiers**

```python
# Identifiers are names used to identify a variable, function, class, module, or any other object.
# They're essentially labels you create to refer to things in your code. identifiers are case-sensitive.

import keyword
# Get the list of all Python keywords
keywords = keyword.kwlist

# Use a for loop with an index (enumerate)
for index, word in enumerate(keywords, start=1):
    print(word, end=', ')
    
    # Check if the index is a multiple of 18
    if index % 18== 0:
        print() # Print a new line
```

```python
# A statement is a command or instruction that the Python interpreter can execute.
# It's a single, complete line of code that performs an action.

x = 10  # This is an assignment statement.
print(x) # This is a function call statement.

if x > 5: # This is a compound statement.
    print("x is greater than 5")
    
# Every expression can be a statement, but not every statement is an expression. 
# For example, if x > 5: is a statement, but it doesn't evaluate to a value.
```

##### string methods

```python
text = "  Hello, world!  "

# Formatting
print(text.strip())       # Output: "Hello, world!"
print(text.upper())       # Output: "  HELLO, WORLD!  "
print(text.replace("o", "a")) # Output: "  Hella, warld!  "

# Query and search
print(text.find("world"))   # Output: 9
print(text.startswith("  Hello")) # Output: True

# Splitting and Joining
words = text.strip().split(", ") # Splits the stripped string by ", "
print(words)              # Output: ['Hello', 'world!']

new_text = " ".join(words)
print(new_text)           # Output: "Hello world!"
```

##### if elif else statement

```python
age = int(input('enter your age'))
if age>18:
    print('you can drive a car')
elif age ==18:
    print('take a test drive')
else:
    print('you are not allowed to drive')
```

```python
marks= int(input('enter your marks: '))
if marks >=90:
    print('your grade is A+')
elif marks >=80 and marks<90:
    print('your grade is A')
elif marks >=70 and marks<80:
    print('your grade is B+')
elif marks >=60 and marks<70:
    print('your grade is B')
elif marks >=50 and marks<60:
    print('your grade is C+')
elif marks >40 and marks<50:
    print('your grade is C')
elif marks <=40:
    print('your grade is F')
```

```python
num= int(input('enter number: '))
if num%2 ==0:
    print(num, 'is even')
else:
    print(num, 'is odd')
```

```python
num= int(input('enter number: '))
if num%2 ==0:
    print(f'{num} is even')
    if num%4 ==0:
        print(f'{num} is divisible by 4')
    else:
        print(f'{num} is not divisible by 4')
else:
    print(f'{num} is odd')
```

### electricity bill using if statement

```python
units= int(input('enter units from electricity bill: '))
value=(units-200)
subsidy=("Subsidy: -200 units")
if units<=200:
    print(units, 'units no charge') 
elif units>200 and units<=300:
    print(subsidy)
    print('Total:',value*5,'\u20B9')
elif units>300:
    print(subsidy)
    print('Total:',value*10,'\u20B9')
```

### task 3

```python
sentence=("the quick brown fox jumps over the lazy dog and the dog runs fast")
word_counts={}
```

```python
words=sentence.split()
```

```python
for word in words:
    if word in word_counts:
        word_counts [word]+=1
    else:
        word_counts [word]=1
```

```python
print(word_counts)
```
