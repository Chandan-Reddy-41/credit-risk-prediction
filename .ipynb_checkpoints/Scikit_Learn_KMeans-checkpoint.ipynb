# Machine learning

```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Dataset
df = pd.read_csv("Dataset_CDS/Mall_Customers.csv")   # dataset path
df.head()
```

#### Select Features (Income vs Spending Score)

```python
# Select columns for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
X.head()
```

#### Feature Scaling 

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### Elbow Method (Find best K)

```python
wcss = []  # within-cluster sum of squares

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_) # inertia_ = WCSS

# Plot Elbow Curve
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='c' )
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal K")
plt.show()
```

#### Train K-Means (Choose K = 5)

```python
kmeans = KMeans(n_clusters=5, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)
```

#### Visualize Clusters

```python
plt.figure(figsize=(8,6))

plt.scatter(X_scaled[y_kmeans == 0, 0], X_scaled[y_kmeans == 0, 1], s=100, label='Cluster 1')
plt.scatter(X_scaled[y_kmeans == 1, 0], X_scaled[y_kmeans == 1, 1], s=100, label='Cluster 2')
plt.scatter(X_scaled[y_kmeans == 2, 0], X_scaled[y_kmeans == 2, 1], s=100, label='Cluster 3')
plt.scatter(X_scaled[y_kmeans == 3, 0], X_scaled[y_kmeans == 3, 1], s=100, label='Cluster 4')
plt.scatter(X_scaled[y_kmeans == 4, 0], X_scaled[y_kmeans == 4, 1], s=100, label='Cluster 5')

# Plot centroids
plt.scatter(kmeans.cluster_centers_[:, 0], 
            kmeans.cluster_centers_[:, 1],
            s=300, marker='X', label='Centroids')

plt.title("Customer Segments - K-Means")
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.legend()
plt.show()
```

#### Silhouette Score (Evaluate Cluster Quality)

Silhouette score tells how well the clusters are separated.

```python
from sklearn.metrics import silhouette_score

# Calculate Silhouette Score
score = silhouette_score(X_scaled, y_kmeans)

print("Silhouette Score:", score)
```

#### Add cluster column to dataset

```python
df['Cluster'] = y_kmeans
df.head()
```

#### Save the model kmeans_model.pkl
Your trained model (you can load it later without training again).

```python
import pickle
# Save the model
with open("kmeans_model.pkl", "wb") as file:
    pickle.dump(kmeans, file)

print("Model saved as: kmeans_model.pkl")
```

#### Save KMeans Predictions file to csv 

```python
# Save predictions
df.to_csv("Mall_Customers_with_clusters.csv", index=False)
print("Predictions saved as: Mall_Customers_with_clusters.csv")

# 3Optional: Show first rows
print(df.head())
```

#### How to Load the Saved Model (Later)

```python
import pickle

# Load the saved KMeans model
with open("kmeans_model.pkl", "rb") as file:
    kmeans_loaded = pickle.load(file)
```

#### Test K-Means Model (predict)
**Imprtant note**: when you train the model with StandardScaler you need to use while testing also. 
because the model is trained on StandardScaler so if you use raw data it will not predict proper it predict all datapoints in one cluster

```python
# ---- Test the model ----
# Example new customer:
# Annual Income = 55k
# Spending Score = 60

new_customer = [[55, 60]]

cluster = kmeans_loaded.predict(new_customer)

print("Predicted Cluster:", cluster[0])
```

```python
test_data = [
    [15, 25],  # low income, low spending
    [80, 10],  # high income, low spending
    [60, 80],  # high income, high spending
    [30, 40],  # mid group
    [90, 90]   # premium customers
] 
# Scale the test data
scaled_test = scaler.transform(test_data)

# Predict clusters
clusters = kmeans_loaded.predict(scaled_test)

# zip(test_data, clusters) pairs each customer with their cluster:
# 'i' is the customer info, 'c' is the cluster number.
for i, c in zip(test_data, clusters):
    print(f"Customer {i}  Cluster {c}") # f-string prints: Customer [15,25]  Cluster 4
```
