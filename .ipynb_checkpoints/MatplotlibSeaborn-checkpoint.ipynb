# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
x= pd.Series([1,2,3,4,5,6,7,7,8])
y= pd.Series([3,4,5,1,2,3,9,10,7])

plt.figure(figsize=(6,4)) #to give the size of the graph
plt.xlabel("This is X-Axis") #to give the title to X axis
plt.ylabel("This is Y-Axis") #to give the title to Y axis
plt.title("Graph") #to give the title to the graph

plt.plot(x,y,linestyle="dashed",marker="^",mfc="red", color="green",markersize=10,linewidth=2) #to make the line chart
plt.show()


```

# Seaborn

```python
np.__version__
```

```python
sns.get_dataset_names() # these are built-in datasets of seaborn library
```

```python
df = sns.load_dataset("iris") #to load the dataset
df
```

```python
df.species.value_counts() #to check the count/frequency of unique values
```

```python
pd.reset_option("display.max_rows",None) #to get the display of the full dataset
df
```

```python
sns.lineplot(x=[0,1,1,1,7,9,2,3,4,5,11],y=[0,9,5,6,9,4,5,4,5,6,11]) #that extra shaded line is CI or significance level or confidence interval by default it is 95
```

```python
sns.lineplot(data=df,x="sepal_length",y="sepal_width")
plt.show()
```

```python
sns.lineplot(data=df,x="sepal_length",y="sepal_width",hue="species",errorbar=("ci",95)) #to make the graph with more than 2 columns or als called multivariate analysis
plt.show()
```

```python
sns.scatterplot(data=df,x="sepal_length",y="sepal_width",hue="species") #scatterplot with multi-variate analysis because the hue is also there 
plt.show()
```

```python
df.species.unique()
```

```python
sns.barplot(x="species",y="petal_width",data=df) #bar chart
plt.show()
```

```python
sns.countplot(x="species",data=df) #to make the count graph of the column in that y axis will give the counting and x axis will hold that categorical data of which the count we want and this is univariate analysis
plt.show()
```

```python
plt.pie(df.species.value_counts(),labels=df.species.unique(),autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
df.info()
```

```python
sns.histplot(x=df.sepal_width,kde=True,hue=df.species)
plt.show()
```

```python
sns.distplot(x=df.sepal_width,kde=True)
plt.show()
```

```python
sns.boxplot(x=df.petal_length)
plt.show()
```

```python
sns.pairplot(df,hue="species")
plt.show()
```

```python
new_data = df.drop("species",axis=1)
new_data
```

```python
new_data.corr() #to take out the correlation among the columns
```
