```python
# Fill missing values in 'Embarked' column with the most frequent value (mode)
# mode()[0] returns the first mode value in case there are multiple modes
# inplace=True modifies the dataframe directly without needing reassignment
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# Fill missing Age values with the median age based on Sex and Pclass groups
# This creates groups by gender and passenger class, then replaces NaN values in Age
# with the median age of the corresponding group
df['Age'] = df.groupby(['Sex', 'Pclass'])['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
# Create a new binary feature 'HasCabin' that equals 1 if the passenger had a cabin (Cabin is not NaN)
# and 0 otherwise. This transforms the presence/absence of cabin information into a numeric feature.
df['HasCabin'] = df['Cabin'].notna().astype(int)
```

```python
# Extract the first character from 'Cabin' column to create a new 'Deck' column
df['Deck'] = df['Cabin'].str[0]
# Replace any NaN values in the 'Deck' column with 'Unknown'
df['Deck'].fillna('Unknown', inplace=True)
```

```python
# Remove the 'Cabin' column from the dataframe as it's likely not needed for analysis
# The inplace=True parameter modifies the dataframe directly without requiring reassignment
df.drop(columns=['Cabin'], inplace=True)
```

```python
# Check the number of missing values in each column of the dataframe
df.isna().sum()
```

```python
def treat_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Option 1 (Capping): Replace outliers with limits
    df[column] = df[column].clip(lower_limit, upper_limit)
    
    # Option 2 (Removing): Uncomment line below to remove rows
    # df = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]
    
    return df
```

```python
def treat_outliers_iqr(df, column):
    # Calculate first quartile (25th percentile)
    Q1 = df[column].quantile(0.25)
    # Calculate third quartile (75th percentile)
    Q3 = df[column].quantile(0.75)
    # Calculate interquartile range (IQR)
    IQR = Q3 - Q1
    
    # Define lower boundary for outliers (Q1 - 1.5*IQR)
    lower_limit = Q1 - 1.5 * IQR
    # Define upper boundary for outliers (Q3 + 1.5*IQR)
    upper_limit = Q3 + 1.5 * IQR
    
    # Option 1 (Capping): Replace outliers with limits
    # Any value below lower_limit will be set to lower_limit
    # Any value above upper_limit will be set to upper_limit
    df[column] = df[column].clip(lower_limit, upper_limit)
    
    # Option 2 (Removing): Uncomment line below to remove rows
    # This alternative approach would remove rows containing outliers instead of capping them
    # df = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]
    
    return df
```

```python
# Apply IQR method to handle outliers in the 'Fare' column
df = treat_outliers_iqr(df, 'Fare')
# Apply IQR method to handle outliers in the 'Age' column
df = treat_outliers_iqr(df, 'Age')
```

```python
# Convert categorical 'Sex' values to numerical: 'male' becomes 0, 'female' becomes 1
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
# Convert 'Embarked' categorical variable to dummy/indicator variables (one-hot encoding)
# Using drop_first=True to avoid multicollinearity (drops the first category)
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
# Convert categorical 'Pclass' variable into dummy/indicator variables (one-hot encoding)
# Creates binary columns for each class (except the first one which is dropped with drop_first=True)
# This transforms 'Pclass' from a single column with values like 1,2,3 into separate binary columns
df = pd.get_dummies(df, columns=['Pclass'], drop_first=True)
```

```python
# Count the frequency of each unique ticket value in the dataset
# This creates a Series where index = ticket values and values = counts
ticket_freq = df['Ticket'].value_counts()
```

```python
df['Ticket'] = df['Ticket'].map(ticket_freq)
```

```python
# Map each ticket value to its frequency in the dataset
# This replaces ticket strings with numeric values representing how common each ticket is
df['Ticket'] = df['Ticket'].map(ticket_freq)
```

## Feature Transformation

### Scaling Transformation (Normalization)
Scaling means changing the range of values so that all features contribute equally to the model.
1. Importing libraries

    - A. StandardScaler  z-score scaling
  
      - Formula: =/

    - B. MinMaxScaler  01 scaling
  
        - Formula: x=xxmin/xmaxxmin

    - C. RobustScaler  ignores outliers
  
        - Formula: Median, IQR (Inter Quartile Range)

2. Fit and Transform

    .fit()  learns scale values (mean, std, min, max)

    .transform()  applies scaling

    .fit_transform()  both at once

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

# sample data
df = pd.DataFrame({
    "Age": [25, 30, 45],
    "Salary": [20000, 50000, 120000]
})

# Standardization
std = StandardScaler()
df_std = std.fit_transform(df)

# MinMax Scaling
mm = MinMaxScaler()
df_mm = mm.fit_transform(df)

# Robust Scaling
rb = RobustScaler()
df_rb = rb.fit_transform(df)

print("Standardization:\n", df_std)
print("MinMax Scaling:\n", df_mm)
print("Robust Scaling:\n", df_rb)
```
