- since loan amount term and credit history have discrete values, we will replace the null values with mode.
- For Loan Amount we will ues the mean/median to replace the values

```python
# Finding Outliers
sns.boxplot(x = df.LoanAmount)
```

- since loan amount has outliers values, we will use median to replace the null values.


```python
df.Loan_Amount_Term.fillna(df.Loan_Amount_Term.mode()[0], inplace = True)
df.Credit_History.fillna(df.Credit_History.mode()[0], inplace = True)
df.LoanAmount.fillna(df.LoanAmount.mode()[0], inplace = True)
```
