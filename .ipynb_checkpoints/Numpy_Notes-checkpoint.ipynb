## 1.1 NumPy Basics - Introduction to Arrays

## What is NumPy?
NumPy (Numerical Python) is a fundamental library for scientific computing in Python. It provides:
- Powerful N-dimensional array objects (ndarray)
- Fast operations on arrays
- Tools for working with numerical data

## Why Use NumPy Arrays?
- **Performance**: NumPy arrays are much faster than Python lists for numerical operations
- **Memory Efficiency**: Uses less memory than Python lists
- **Convenience**: Provides many mathematical functions and operations

## Key Concept: ndarray
The core of NumPy is the ndarray (n-dimensional array) - a homogeneous collection of elements of the same data type.

## Importing NumPy Library
Before we can use NumPy, we need to import it into our Python program. By convention, we import NumPy with the alias 'np' for convenience.

```python
# Import the entire NumPy library and assign it the alias 'np'
import numpy as np
```

## 2.1 Creating a 1-Dimensional Array (Vector)
The simplest array is a 1D array, which is like a standard Python list but with the NumPy speed and features.

```python
# Create a standard Python list of stock prices
stock_prices_list = [150.5, 151.2, 149.8, 152.0, 153.1]

# Convert the Python list into a NumPy array
# This is our first 1-Dimensional (1D) array, or a 'vector'
stock_prices_array = np.array(stock_prices_list) 

# Print the resulting array
print(stock_prices_array)

# Check the type of the new object
print(type(stock_prices_array))
```

## 2.2 Array Attributes

## Attribute | Description

**.ndim** | The number of dimensions (axes) of the array (rank)

**.shape** | A tuple indicating the size of the array in each dimension

**.size** | The total number of elements in the array

**.dtype** | The data type of the elements in the array (e.g., int64, float64)


```python
# Let's check the attributes of our stock_prices_array
print(f"Number of dimensions (.ndim): {stock_prices_array.ndim}")
print(f"Shape of the array (.shape): {stock_prices_array.shape}")
print(f"Total number of elements (.size): {stock_prices_array.size}")
print(f"Data type of elements (.dtype): {stock_prices_array.dtype}")
```

## 3.1 Creating a 2D Array from Nested Lists 

Concept Explanation: A 2D array is created by passing a list of lists to the **np.array()** function, where each inner list represents a row in the matrix.

```python
# Create a nested Python list (list of lists)
# Inner lists are the rows
# Columns are the metrics

Data_list = [[25, 65, 10],[28, 60, 15],[22, 70, 5]]

# Convert the nested list into a 2D NumPy array (a 'matrix')
Data_matrix = np.array(Data_list)

# Print the matrix
print("Data Matrix:")
print(Data_matrix)
```

## 3.2 Checking 2D array attribute

```python
# Let's check the attributes of our 2D array
print(f"Number of dimensions (.ndim): {Data_matrix.ndim}")
print(f"Shape of the array (.shape): {Data_matrix.shape}")
print(f"Total number of elements (.size): {Data_matrix.size}")
print(f"Data type (.dtype): {Data_matrix.dtype}")
```

**np.zeros(shape)** | Creates an array filled with zeros. | np.zeros((2,Â 4))

**np.ones(shape)** | Creates an array filled with ones. | np.ones(3)

**np.arange(start,Â stop,Â step)** | Creates an array with evenly spaced values within a given interval. (Similar to Python's range()). | np.arange(0,Â 10,Â 2)

**np.linspace(start,Â stop,Â num)** | Creates an array of a specified number of evenly spaced values over an interval. | np.linspace(0,Â 1,Â 5)

## 5. Universal Array Operations (UFuncs) 
This is where NumPy's power really shines! Unlike standard Python lists, you can perform mathematical operations on entire NumPy arrays simultaneously, without using slow for loops. These are called Universal Functions or UFuncs.

## 5.1 Element-Wise Arithmetic
When you perform a standard arithmetic operation (like +, -, *, /), NumPy applies that operation to every single element in the array.

## 5.2 Arithmetic Between Two Arrays
You can also perform element-wise arithmetic between two arrays, provided they have the exact same **.shape**.

Real-World Example: You have the expected sales targets and the actual sales achieved. You want to calculate the difference (the variance).

```python
# Sales targets for the week
targets = np.array([16, 19, 15, 20, 21])

# Actual sales achieved 
actual_sales = np.array([15.2, 18.5, 14.0, 20.1, 19.8])

print(f"Targets:      {targets}")
print(f"Actual Sales: {actual_sales}")

# Calculate the variance: Actual - Target (element-by-element subtraction)
variance = actual_sales - targets

print(f"Variance (Actual - Target): {variance}")

# Check if the arrays have the same shape - CRUCIAL for element-wise operations!
print(f"\nDo the shapes match? {targets.shape == actual_sales.shape}")
```

## 5.3 Aggregate Functions (Statistical Operations)
Data science relies heavily on summarizing data. NumPy provides highly optimized functions for calculating statistics across the entire array or along specific dimensions.

Function | Description 

**.sum()** | Calculates the sum of all elements.

**.mean()** | Calculates the average (arithmetic mean) of all elements.

**.std()** | Calculates the standard deviation.

**.min()/.max()** | Finds the minimum/maximum value.

**.argmin()/.argmax()** | Finds the index of the minimum/maximum value.

**.sort()** | Sorts the list/array

**.argsort()** | Finds an index in the original array (axis=1 sorts rows axis= 2 sorts columns)

```python
# Using our actual_sales array
actual_sales = np.array([15.2, 18.5, 14.0, 20.1, 19.8])

print(f"Total Sales (Sum): {actual_sales.sum()}")
print(f"Average Daily Sales (Mean): {actual_sales.mean():.2f}") # Format to 2 decimal places(:.2f)
print(f"Std Sales Day (Std): {actual_sales.std()}")
print(f"Lowest Sales Day (Min): {actual_sales.min()}")
print(f"Highest Sales Day (Max): {actual_sales.max()}")

# Find the index (position) of the highest sales day
print(f"Index of lowest Sales Day (Argmin): {actual_sales.argmin()}")
print(f"Index of Highest Sales Day (Argmax): {actual_sales.argmax()}")

# Finds an indices in the original array
print(f"Index of orginal array (sort): {np.sort(actual_sales)}")
print(f"Index of orginal array (Argsort): {actual_sales.argsort()}")
```

## 6. Array Indexing and Slicing âœ‚ï¸
Accessing specific data points or subsets of an array is one of the most frequent tasks in data science. NumPy uses a concept similar to Python list indexing and slicing but extends it powerfully to multiple dimensions.

## 6.1 1D Array Indexing (Like Python Lists)
Concept Explanation: For 1D arrays, we use square brackets **[ ]** and a single integer index (starting from 0) to access an element. Negative indexing works the same as in Python lists (e.g., **-1** is the last element).

```python
traffic_week_1 = np.array([10, 12, 11, 15, 16, 8, 5])
# Indices:                  0   1   2   3   4  5  6
# Negative Indices:        -7  -6  -5  -4  -3 -2 -1

# Get the traffic for the first day (index 0)
first_day_traffic = traffic_week_1[0] 

# Get the traffic for the last day (index -1)
last_day_traffic = traffic_week_1[-1] 

print(f"Traffic for the first day: {first_day_traffic}")
print(f"Traffic for the last day: {last_day_traffic}")
```

## 6.2 1D Array Slicing (Subsetting)
Concept Explanation: Slicing allows you to extract a subset (a range of elements) from the array using the colon notation: **[start:stop:step]**}. The **start** index is inclusive. The **stop** index is exclusive (meaning it stops before this index). The **step** is optional (default is 1).

```python
# traffic_week_1: [10, 12, 11, 15, 16, 8, 5]

# Get the traffic from the 2nd day (index 1) up to the 5th day (index 4 is exclusive)
midweek_traffic = traffic_week_1[1:5]
print(f"Traffic from Day 2 to Day 5: {midweek_traffic}") # Indices 1, 2, 3, 4

# Get all traffic values from the 4th day (index 3) to the end
traffic_to_end = traffic_week_1[3:]
print(f"Traffic from Day 4 onwards: {traffic_to_end}")

# Get every other day's traffic starting from the first day
every_other_day = traffic_week_1[::2]
print(f"Every other day traffic: {every_other_day}")
```

## 6.3 2D Array Indexing and Slicing (Rows and Columns)
This is the most crucial part for matrices (2D arrays). You now need to specify both the row and the column.

Concept Explanation: The indexing format is array [**row_index, column_index**].

Real-World Example: We'll use a 3Ã—3 matrix representing weather data (Rows = Day, Columns = Metric: Temp, Humidity, Wind).

```python
weather_matrix = np.array([[25, 65, 10],
                           [28, 60, 15],
                           [22, 70, 5]])

# 1. Access a single element (e.g., Humidity on Day 2)
# Row 1, Column 1
day2_humidity = weather_matrix[1, 1] 
print(f"Humidity on Day 2: {day2_humidity}")

# 2. Access an entire Row (e.g., all data for Day 3)
# Use a single index for the row, and a colon (:) for all columns
day3_data = weather_matrix[2, :] 
print(f"Data for Day 3 (Row 2): {day3_data}")

# 3. Access an entire Column (e.g., Temperature for all days)
# Use a colon (:) for all rows, and a single index for the column
all_days_temp = weather_matrix[:, 0] 
print(f"Temperature for all days (Col 0): {all_days_temp}")
```

## 7. Advanced Indexing ðŸ’¡
While basic slicing is useful, real-world data science often requires selecting data based on a condition (like "sales over 100") or selecting non-contiguous data points. This is where Boolean Indexing and Fancy Indexing come in.

## 7.1 Boolean Indexing (Filtering Data)
Concept Explanation: Boolean indexing uses a Boolean array (an array of **True**/**False** values) to select elements. Only the elements where the Boolean array is **True** are kept in the resulting array. This is the most common way to filter data in NumPy. 

Real-World Example: Imagine filtering a list of product prices to find all products that are currently on sale (price less than 20).

```python
# Array of product prices
prices = np.array([15.50, 22.99, 19.99, 35.00, 12.00])

# Step 1: Create the Boolean array (the mask)
# This compares every element to the condition (price < 20)
is_on_sale = prices < 20
print(f"1. Boolean Mask: {is_on_sale}")

# Step 2: Use the Boolean array to filter the original array
sale_prices = prices[is_on_sale] 
print(f"2. Sale Prices: {sale_prices}")

# You can do this in one line:
high_prices = prices[prices > 30] 
print(f"3. Prices > 30: {high_prices}")
```

Explanation: The Boolean mask tells NumPy exactly which positions to include. The element at index 1 is $\texttt{False}$, so the price $\texttt{22.99}$ is excluded.

## Combining Conditions
You can use the logical operators **&** (AND) and **|** (OR) to combine multiple conditions. You must wrap each condition in parentheses. 

use symbols (**&**) AND/(**|**) OR

```python
# Find prices between $15 and $30 (inclusive)
# Condition 1: greater than or equal to 15 (prices >= 15)
# AND
# Condition 2: less than or equal to 30 (prices <= 30)

mid_range_prices = prices[(prices >= 15) & (prices <= 30)]
print(f"Mid-Range Prices ($15-$30): {mid_range_prices}")
```

## 7.2 Fancy Indexing (Selecting Specific Indices)
Concept Explanation: Fancy indexing involves passing an array or list of integer indices to select specific, non-sequential elements.

Real-World Example: Imagine you have a list of monthly sales and you only want to check the sales from the 1st, 5th, and 10th month.

```python
monthly_sales = np.arange(100, 130, 3) 
# [100 103 106 109 112 115 118 121 124 127] (10 months of sales)
print(f"All Monthly Sales: {monthly_sales}")

# We want sales from index 0, index 4, and index 9
target_indices = [0, 4, 9]

# Use the list of indices to select the elements
selected_sales = monthly_sales[target_indices]
print(f"Sales from Month 1, 5, and 10: {selected_sales}")
```

# 8. Broadcasting ðŸ“¢
Broadcasting is the mechanism NumPy uses to allow arrays of different shapes to perform arithmetic operations. 

It's what made operations like **daily\_sales + 100** possible earlier.

## 8.1 The Core Idea Concept 
Explanation: When performing an operation between two arrays (or an array and a single number/scalar), NumPy tries to "stretch" the smaller array across the larger array so that they have compatible shapes. The operation is then performed element-wise.

Think of it like this: 
Operation **array** + **scalar** NumPy's internal thought process: "I will treat the scalar as an array of the same shape as the other array, and every element in this new 'stretched' array is the value of the scalar."

```python
data = np.array([1, 2, 3, 4])
multiplier = 10 

# NumPy conceptually 'broadcasts' 10 to [10, 10, 10, 10]
result_1d = data * multiplier 
print(f"Data: {data}")
print(f"Result (Data * 10): {result_1d}")
```

## 8.2 Broadcasting with 2D Arrays
Broadcasting becomes particularly useful and a bit more complex with 2D arrays (matrices). For two arrays to be compatible for broadcasting, their dimensions must satisfy one of two rules, starting from the trailing (rightmost) dimension:

The dimensions must be equal.

One of the dimensions must be 1.

```python
# A 3x2 Matrix (3 products, 2 regions)
sales_matrix = np.array([[100, 150],
                         [200, 250],
                         [50, 75]])
# Shape: (3, 2)

# A 1D array representing a discount rate for each region
# Region 1 gets 10% off (0.9 multiplier), Region 2 gets 5% off (0.95 multiplier)
discount_rate = np.array([0.9, 0.95]) 

print("Sales Matrix:")
print(sales_matrix)
print(f"\nDiscount Rates (Shape: {discount_rate.shape}): {discount_rate}")

# When we multiply, the 1D array (shape 2) is broadcast across the rows of the 3x2 matrix.
# [0.9, 0.95] is effectively copied three times:
#  [[0.9, 0.95],
#  [0.9, 0.95],
#  [0.9, 0.95]]
final_sales = sales_matrix * discount_rate

print("\nFinal Sales (Sales * Discount Rate):")
print(final_sales)
```

Explanation: The rightmost dimension of $\texttt{sales\_matrix}$ is 2, and the dimension of $\texttt{discount\_rate}$ is 2. Since they are equal, NumPy broadcasts the smaller array (the rates) along the rows of the larger array, applying the 0.9 to the first column and 0.95 to the second column for all rows.

## 9. Array Reshaping ðŸ“
Reshaping is critical in data science because models and functions often require input arrays to have a very specific shape (e.g., a function might require a 2D array, even if your data is 1D).

## 9.1 The .reshape() Method 
Concept Explanation: The **.reshape()** method gives a new shape to an array without changing the data. It requires the new shape to have the same total number of elements as the original array.

Real-World Example: Converting a list of 12 monthly sales figures into a **4 \times 3** matrix (4 quarters, 3 months per quarter).

```python
# 1D array of 12 monthly sales figures
monthly_sales = np.arange(101, 113) # Creates [101, 102, ..., 112]
print(f"Original 1D array (Size {monthly_sales.size}): {monthly_sales}")

# Reshape the 1D array (size 12) into a 2D array of 4 rows and 3 columns
quarterly_sales = monthly_sales.reshape(4, 3) 

print("\nReshaped 4x3 Quarterly Sales Matrix:")
print(quarterly_sales)
print(f"New Shape: {quarterly_sales.shape}")
```

Important: The new shape must maintain the original size (12 elements in this case: $4 \times 3 = 12$).

## 9.2 The Power of **-1** (Let NumPy Decide)

Concept Explanation: You can use **-1** in the **.reshape()** method to tell NumPy to automatically calculate one dimension based on the size of the array and the other dimension you provide.

Example: You have 12 items and you know you want 3 columns, but you don't want to calculate how many rows that is (**12 / 3 = 4**).

```python
# Original array (size 12)
Original_array= np.array([101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111, 112])

# Reshape into a matrix with 3 columns, let NumPy calculate the rows (-1)
three_column_matrix = Original_array.reshape(-1, 3) 
print("\nReshaped with -1 (3 columns, 4 rows calculated):")
print(three_column_matrix)
print(f"New Shape: {three_column_matrix.shape}")

# Reshape into a matrix with 6 rows, let NumPy calculate the columns (-1)
six_row_matrix = Original_array.reshape(6, -1)
print("\nReshaped with -1 (6 rows, 2 columns calculated):")
print(six_row_matrix)
print(f"New Shape: {six_row_matrix.shape}")
```

## 9.3 Flattening an Array
Concept Explanation: The **.ravel()** or **.flatten()** methods convert any multi-dimensional array back into a 1-dimensional array.

**.ravel()**: If you modify the new 1D array, the original array is modified (if it's a view).

**.flatten()**: If you modify the new 1D array, the original array is NOT modified (since it's a copy).

```python
quarterly_sales=np.array([[101, 102, 103],
                          [104, 105, 106],
                          [107, 108, 109],
                          [110, 111, 112]])
print(quarterly_sales)

ravel_array = quarterly_sales.ravel() # Using .ravel() - creates a VIEW
print(f"\nRavel Array (1D): {ravel_array}")
print(f"New Shape: {ravel_array.shape}")
print(ravel_array.ndim)


flatten_array = quarterly_sales.flatten() # Using .flatten() - creates a COPY
print(f"\nFlatten Array (1D): {flatten_array}")
print(f"New Shape: {flatten_array.shape}")
print(flatten_array.ndim)
```

```python
# Modify the first element of the raveled array
ravel_array[0] = 99 
print(ravel_array)
print(f"Quarterly_sales Array after Ravel modification:\n{quarterly_sales}")

# Reset Quarterly sales array
quarterly_sales=np.array([[101, 102, 103],
                          [104, 105, 106],
                          [107, 108, 109],
                          [110, 111, 112]])
print("\n")
# Modify the first element of the flattened array
flatten_array[0] = 88
print(flatten_array)
print(f"Quarterly_sales Array after Flatten modification:\n{quarterly_sales}")
```
