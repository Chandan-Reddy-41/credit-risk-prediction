```python
sales_data = {
    'Region': ['North', 'South', 'North', 'West', 'South', 'West'],
    'Salesperson': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob', 'David'],
    'Revenue': [1500, 2200, 1800, 3000, 1900, 2500],
    'Items_Sold': [10, 15, 12, 20, 14, 18]
}
df_sales = pd.DataFrame(sales_data)

print("Original Sales Data:")
print(df_sales)

# 1. Group the data by the 'Region' column
by_region = df_sales.groupby('Region')

# 2. Apply an aggregation function (e.g., calculate the SUM of Revenue and Items_Sold for each region)
# Only numerical columns are summed.
regional_totals = by_region.sum() 

print(f"\nRegional Sales Totals (Sum):\n{regional_totals}")
```

```python
# Calculate the average (mean) Revenue and Items_Sold per region
regional_averages = df_sales.groupby('Region').mean(numeric_only=True)

print(f"\nRegional Sales Averages (Mean):\n{regional_averages}")
```

```python
sales_data = {
    'Region': ['North', 'South', 'North', 'West', 'South', 'West'],
    'Salesperson': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob', 'David'],
    'Revenue': [1500, 2200, 1800, 3000, 1900, 2500],
    'Items_Sold': [10, 15, 12, 20, 14, 18]
}
df_sales = pd.DataFrame(sales_data)

# 1. Group by Salesperson and calculate total Items_Sold
total_items = df_sales.groupby('Salesperson')['Items_Sold'].sum()
print("1. Total Items Sold by Salesperson:")
print(total_items)

# 2. Calculate the average Revenue earned by each Salesperson
avg_revenue = df_sales.groupby('Salesperson')['Revenue'].mean()
print("\n2. Average Revenue by Salesperson:")
print(avg_revenue)

# 3. Group by Region and use .describe() on Revenue
regional_revenue_summary = df_sales.groupby('Region')['Revenue'].describe()
print("\n3. Revenue Summary by Region (.describe()):")
print(regional_revenue_summary)
```

```python
# Create a simple DataFrame
df_scores = pd.DataFrame({'Score': [85, 92, 78, 95, 60]})

# Function to assign a letter grade
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'F'

# Apply the function to the 'Score' column (which is a Series)
df_scores['Grade'] = df_scores['Score'].apply(get_grade)

print("\nApplying Custom Function (.apply()):")
print(df_scores)
```

```python
df_scores['Pass_Fail'] = df_scores['Grade'].map({'A': 'Pass', 'B': 'Pass', 'C': 'Pass', 'F': 'Fail'})
print("\nApplying Dictionary Mapping (.map()):")
print(df_scores)
```

```python
# Function to assign a letter grade
def get_Result(Grade):
    if Grade in {'A','B','C'}:
        return 'Pass'
    else:
        return 'Fail'

# Apply the function to the 'Score' column (which is a Series)
df_scores['Result'] = df_scores['Grade'].apply(get_Result)

print("\nApplying Custom Function (.apply()):")
print(df_scores)
```

```python
# Sample data with Region, Product, and Sales
df_data = pd.DataFrame({
    'Region': ['East', 'East', 'West', 'West', 'East'],
    'Product': ['A', 'B', 'A', 'B', 'A'],
    'Sales': [100, np.nan, 120, 200, 130]
})

# Pivot: Summarize total Sales by Region AND Product
sales_pivot = df_data.pivot_table(
    index='Region', 
    columns='Product', 
    values='Sales', 
    aggfunc='sum'
)

print("\nPivot Table (Region vs. Product Sales):")
print(sales_pivot)
```

```python
# Count how many times each Product appears in each Region
frequency_table = pd.crosstab(df_data['Region'], df_data['Product'])
print("\nCrosstab (Frequency Table):")
print(frequency_table)
```

```python
# Group by both Region AND Product, and calculate the mean Sales
multi_group = df_data.groupby(['Region', 'Product'])['Sales'].mean()

print("\nSeries with MultiIndex (Region, Product):")
print(multi_group)
print(f"Index Type: {multi_group.index}")
```

```python
import requests
from io import StringIO

# 1. Define custom headers to mimic a common web browser
headers = {
    # This is the most critical header to spoof the server
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36',
    # Optionally add a referer to look more legitimate
    'Referer': 'https://www.google.com/', 
    # Optional: Accept compressed data
    'Accept-Encoding': 'gzip, deflate, br', 
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'Connection': 'keep-alive',
}

url = 'https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)'

try:
    # 2. Make the request with the headers
    response = requests.get(url, headers=headers)
    
    # 3. Check the status code
    if response.status_code == 200:
        # Pass the text content to pandas
        df_list = pd.read_html(StringIO(response.text))
        
        # Select your desired table
        df = df_list[0] 
        print(" Data successfully loaded after adding User-Agent header!")
        print(df.head())
        
    else:
        print(f" Failed to load data. Status code: {response.status_code}")
        
except requests.exceptions.RequestException as e:
    print(f"An error occurred during the request: {e}")
```
